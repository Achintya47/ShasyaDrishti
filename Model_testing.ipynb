```python
import pandas as pd
from sklearn.metrics import r2_score
import joblib
from sklearn.preprocessing import LabelEncoder
```

```python
model1=joblib.load("Gradient_Boosting.pkl")
best_model=joblib.load("Best_XGB_Regressor.pkl")
```

```python
df=pd.read_csv("final_merged.csv")
df.head()
```

```python
'''Encoding the categorical columns'''
le=LabelEncoder()
df['Crop']=le.fit_transform(df['Crop']) 
df['Season']=le.fit_transform(df['Season'])
df['Pollutants']=df['so2'] + df['no2'] + df['rspm'] + df['spm'] + df['pm2_5']
df.head()
```

```python
y=df['Production']
X = df.drop(['Production', 'State_Name', 'District_Name', 'Crop_Year'], axis=1)
```

```python
y_pred1=model1.predict(X)
y_pred2=best_model.predict(X)
```

```python
score1=r2_score(y,y_pred1)
score2=r2_score(y,y_pred2)
```
